```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data=pd.read_csv("C:/Users/priya/OneDrive/Desktop/python datasets/Telco customer churn.csv")
a=sns.countplot(data=data,x="Churn")
a.bar_label(a.containers[0])
plt.title("Count of customers by Churn")
plt.show()
```

```python
a=data.groupby("Churn").agg({"Churn":"count"})
print(a)
plt.pie(x=a["Churn"],labels=a.index,autopct="%.2f%%")
plt.title("Percentage of Churned Customers",fontsize=12)
plt.show()
```

```python
sns.countplot(data=data,x="gender",hue="Churn")
plt.title("Churn by gender")
plt.show()
```

```python
a=data.groupby("gender").agg({"gender":"count"})
print(a)
plt.pie(x=a["gender"],labels=a.index,autopct="%.2f%%")
plt.title("Percentage of Male and Female customers")
plt.show()
```

```python
def conv(value):
    if value==1:
        return "yes"
    else: 
        return "no"

data["SeniorCitizen"]=data["SeniorCitizen"].apply(conv)
plt.figure(figsize=(3,3))
sns.countplot(data=data,x="SeniorCitizen",hue="Churn")
plt.title("Churn by SeniorCitizen")
plt.show()
```

```python
# Create count table
ct = pd.crosstab(data['SeniorCitizen'], data['Churn'])

# Convert to percentages of total in each SeniorCitizen group
pct = ct.div(ct.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
ax = pct.plot(
    kind='bar',
    stacked=True,
    figsize=(6,4)
)

# Add percentage labels
for container in ax.containers:
    ax.bar_label(
        container,
        fmt='%.1f%%',
        label_type='center'
    )

plt.title("Churn by Senior Citizen (%)")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage")
plt.legend(title="Churn")
plt.tight_layout()
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

fig, axes = plt.subplots(3, 3, figsize=(18, 12))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.countplot(
        data=data,
        x=col,
        hue='Churn',
        ax=axes[i]
    )

    axes[i].set_title(f'Count plot of {col}')
    axes[i].tick_params(axis='x', rotation=30)

    # Show legend only on first subplot
    if i != 0:
        axes[i].get_legend().remove()

# Add a single legend for the whole figure
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, title='Churn', loc='upper right')

plt.tight_layout()
plt.show()
#do by subplot for plot the count plot of all rests columns together
```

#Customers without OnlineSecurity and TechSupport have substantially higher churn than those who subscribe to these services, making them strong indicators of customer retention.
#Among InternetService types, Fiber Optic users show the highest churn counts, while customers with no internet service have the lowest churn.
#Services such as OnlineBackup, DeviceProtection, StreamingTV, and StreamingMovies show moderate differences, with churn generally lower among customers who use these services.
#Overall, customers who subscribe to additional support and security-related services tend to churn less, suggesting that value-added services improve customer loyalty.
